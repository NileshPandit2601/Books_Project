```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path = r'C:\Users\niles\Desktop\books.csv' 
df = pd.read_csv(path)
print(df.head())
df.info()
```

```python
df.describe(include = 'all')
```

#### The dataset consist of 1000 books and all unique.
#### The dataset only contains the books which are in stock.
#### The price ranges from 10 - 59.99 pounds.
#### The range of ratings is 1 to 5.
#### The average price of books is 35 pound with average ratings of 2.923.

```python
rating_counts = df['Rating'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(rating_counts.index, rating_counts.values)
plt.xlabel('Rating (Stars)')
plt.ylabel('Number of Books')
plt.title('Number of Books for Each Rating')
plt.xticks([1, 2, 3, 4, 5])
plt.tight_layout()
plt.show()
```

#### The most rating given by customers is 1 and the lowest is 4.
#### The lowest rating consist of highest number of books.
#### While 2,3 and 5 have similiar number of books.

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Price', bins=20)

plt.title('Distribution of Book Prices')
plt.xlabel('Price ()')
plt.ylabel('Number of Books')
plt.tight_layout()
plt.show()
```

#### The 20-22 pound price book has the highest frequency
#### The 58-60 pound price book has the lowest frequency

```python
stock_counts = df['Availability'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(stock_counts, labels=stock_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Books In Stock vs. Out of Stock')
plt.tight_layout()
plt.show()
```

#### Every book in dataset is in stock.
#### The dataset only contains the book which are available in stock
